```python
def print_var_info(o, name=None, no_print_value=False):
    if no_print_value:
        if name:
            print("%s id(%s) - %r" % (name, id(o), type(o), ))
        else:
            print("id(%s) - %r" % (id(o), type(o), ))
    else:
        if name:
            print("%s id(%s) = %r - %r" % (name, id(o), o, type(o), ))
        else:
            print("id(%s) = %r - %r" % (id(o), o, type(o), ))
```

```python
from typing import Iterator, Generator, Iterable


def check_types(o, name=None):
    name = name or "object"
    for t in (Iterable, Iterator, Generator):
        print("Is %r %s: %s" % (name, str(t), isinstance(o, t)))
    
    for m in ("__iter__", "__next__"):
        print("Is %r has attribute %s: %s" % (name, str(m), hasattr(o, m)))
```

# Iterators

```python
class Squares:
    def __init__(self, start, stop):
        self.start = start
        self.stop = stop
    
    def __iter__(self): 
        return self
    
    def __next__(self):
        if self.start > self.stop:
            raise StopIteration
        
        current = self.start ** 2
        self.start += 1

        return current
```

```python
sq = Squares(1, 3)
```

```python
dir(sq)
```

```python
for item in Squares(1, 3):
    print(item)
```

# Generators

```python
import time

def squares(start, stop):
    while start <= stop:
        yield start ** 2
        start += 1
```

```python
for item in squares(1, 3):
    print(item)
```

```python
sq_gen = squares(1, 3)
```

```python
sq_gen.__class__
```

```python
dir(sq_gen)
```

```python
sq1_gen = squares(1, 3)

first_value = next(sq1_gen)

if first_value < 2:
    print(first_value)
    for x in sq1_gen:
        print(x)
else:
    print(first_value)
```

```python
def skip_line(file, skip=0):
    with open(file, "rb") as fp:
        try:
            for _ in range(skip):
                next(fp)
        except StopIteration:
            return
        
        yield from fp
#         for item in fp:
#             yield fp
```

```python
for line in skip_line(sample_file, 0):
    print(line)
```
