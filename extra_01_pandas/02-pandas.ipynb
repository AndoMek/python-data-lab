```python
import os
import csv

csv.field_size_limit(2**31 - 1)  # Workaround to avoid _csv.Error: field larger than field limit (131072)


def csv2dict(csv_filename, columns=None, delimiter=",", quotechar='"'):
    # Just for example for csv you can use DataFrame.from_csv method

    filename = os.path.expanduser(csv_filename)
    filename = os.path.expandvars(filename)

    with open(filename, "rt") as csv_file:
        yield from csv.DictReader(
            csv_file, 
            fieldnames=columns,
            delimiter=delimiter,
            quotechar=quotechar,
        )
```

```python
import datacompy


compare = datacompy.Compare(
    df_events,
    df_events_db,
    join_columns="event_id",
    df1_name='in_csv',
    df2_name='in_db'
    )
compare.matches(ignore_extra_columns=False)

print(compare.report())
```
